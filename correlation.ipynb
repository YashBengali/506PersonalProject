```python
import pandas as pd
from scipy.stats import pearsonr


def read_tweets_with_sent():
    df = pd.read_csv('tweets.csv', nrows=48978)
    # print(df.head())
    # print(df.iloc[-1])
    # print(df.iloc[0])
    # print(df.isnull().values.any())
    return df
tweets = read_tweets_with_sent()
# tweets.dropna()
# print(tweets.head())

def read_sent():
    df = pd.read_csv('sentiments_copy.csv')
    return df

def read_yahoo():
    df = pd.read_csv('AAPL.csv', nrows=253)
    # print(df)
    return df
yahoo = read_yahoo()
# print(yahoo.head())
yahoo['change'] = yahoo.apply (lambda row: (row['Close'] - row['Open']), axis=1)
yahoo = yahoo.drop(['Open', 'Close', 'Low', 'High', 'Adj Close', 'Volume'], axis=1)
# print(yahoo.head())
yahoo['date'] = yahoo.apply(lambda x: str(x['date']), axis=1)
# print(yahoo['date'])

sent = read_sent()
# print(sent.head())
combined = pd.concat([tweets,sent], axis=1)
# print(combined.head())
print(sent.isnull().values.any())
print(tweets.isnull().values.any())
print(combined.isnull().values.any())

full_combined = combined.copy()

combined = combined.drop(['followers', 'likes', 'tweet', 'MAGNITUDE'], axis=1)

combined['date'] = combined['date'].apply(lambda x: x if type(x) == type(1.0) else str(x[:10]))
# print(combined['date'])
# print(combined.head())

grouped = combined.groupby(by=['date']).mean()
# print(combined.iloc[0]['date'])
# print(combined.iloc[15860])
# print(combined.iloc[0] == combined.iloc[15860])
# print(grouped)

both = yahoo.merge(right=grouped, how='right', left_on='date', right_on ='date')

both = both.dropna()
print(both)

both = both.drop(['date'], axis=1)
both = both[both.columns[::-1]]
both = both.dropna()
print(both.corr())
# yahoo = yahoo.iloc[236:248]
# print(yahoo)
pearsonr(both['SCORE'], both['change'])
```
