### Read csv and preprocess

```python
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer


tweets = pd.read_csv("tweets.csv")
date = tweets.CREATED.str.split(expand=True)
tweets["DATE"] = date[0]
tweets.drop(columns=['CREATED'])
tweets.sort_values(by=['DATE'])
tweets = tweets[tweets.DATE != 'date']
tweets = tweets[tweets.DATE != '=======']
print(tweets.DATE.min())
print(tweets.DATE.max())


analyzer = SentimentIntensityAnalyzer()
tweets['compound'] = [analyzer.polarity_scores(x)['compound'] for x in tweets['CONTENT']]
tweets['SENTIMENT'] = tweets.compound /tweets.compound.abs().max()
tweets.drop(columns=['compound'])
print(tweets)
```

### Read Apple stock data

```python
stock_data = pd.read_csv("AAPL.csv")
print(stock_data)
```

```python
grouped_df = tweets.groupby(by = ["DATE"])
sentiment = []
date=[]
for name,group in grouped_df:
    date.append(name)
    sentiment.append(group.SENTIMENT.mean())
stock_df = pd.DataFrame({"Date": date, "Mean_Sentiment": sentiment})
stock_df
```

```python
import numpy as np
stock_data["increased"] = np.where(stock_data["Close"]>stock_data["Open"], 1, 0)
stock_data["change_value"] = stock_data["Close"]-stock_data["Open"]
```

```python
x = pd.merge(stock_data, stock_df, on="Date")
x
```

```python
print(x["change_value"].corr(x["Mean_Sentiment"]))
print(x["increased"].corr(x["Mean_Sentiment"]))
```
